## Normalization
<pre><font size=3>
Normalization is a technique often applied as part of data preparation for machine learning. The goal of normalization is to change the values of numeric columns in the dataset to use a common scale, without distorting differences in the ranges of values or losing information. Normalization is also required for some algorithms to model the data correctly.

For example, assume your input dataset contains one column with values ranging from 0 to 1, and another column with values ranging from 10,000 to 100,000. The great difference in the scale of the numbers could cause problems when you attempt to combine the values as features during modeling.

Normalization avoids these problems by creating new values that maintain the general distribution and ratios in the source data, while keeping values within a scale applied across all numeric columns used in the model.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_values_y = np.random.random(1000)*1000
random_values_x = np.random.random(1000)*10
plt.scatter(random_values_x,random_values_y)
```

```python
plt.hist(random_values_y,alpha=0.5,edgecolor='black')
plt.show()
```

```python
random_values_y2 = np.random.randint(700,10000,1000)/10
plt.hist(random_values_y2,alpha=0.5, color='steelblue', edgecolor='black')
plt.show()
```

```python
print("Mean : {}\nVariance : {}\nSTD : {}".format(random_values_y2.mean(),random_values_y2.var(),random_values_y2.std()) )
```

```python
# Normalizing
y2_mean = random_values_y2.mean()
z = (random_values_y2-y2_mean)/random_values_y2.std()
```

```python
plt.hist(z,alpha=0.7, edgecolor='black')
plt.show()
```

```python
print("Mean : {}\nVariance : {}\nSTD : {}".format(z.mean(), z.var(), z.std()) ) 
```

```python
print("Mean : {}\nVariance : {}\nSTD : {}".format(round(z.mean(),3), round(z.var(),3), round(z.std(),3)) )
```

```python
from sklearn import preprocessing
```

```python
sklearn_norm_y = preprocessing.normalize([random_values_y2])
print("Mean : {}\nVariance : {}\nSTD : {}".format(round(sklearn_norm_y.mean(),3), 
                                                  round(sklearn_norm_y.var(),3), 
                                                  round(sklearn_norm_y.std(),3)) )
```

# BAYES THEOREM
Formula For Bayes' Theorem <br>


- Joint Probability
<br>
$ P(A and B and C) = P(A) * P(B) * P(C)$
<br><br>
- Disjoint Probability
<br>
$ P(A or B or C) = P(A) + P(B) + P(C)$
<br><br>
- Conditional Probability
<br>
$P(A|B)=\frac{P(A and B)}{P(A)}$
<br><br><br>
- BAYES
<br>
$P(A|B)= \frac{P(A⋂B)}{P(B)} = \frac{P(A)*P(B|A)}{P(B)} $
<br><br>
<pre>
P(A)= The probability of A occurring
P(B)= The probability of B occurring
P(A∣B)=The probability of A given B
P(B∣A)= The probability of B given A
P(A⋂B))= The probability of both A and B occurring
